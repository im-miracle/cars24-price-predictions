```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.linear_model import Ridge
import numpy as np
from functools import partial

# Example custom scoring function
def customFunc(y_true, y_pred, X_train):
    n, p = X_train.shape
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    
    print("Adjusted R^2: ", adjusted_r2)
    return adjusted_r2

# Freeze X_train using functools.partial
custom_scorer = make_scorer(partial(customFunc, X_train=X_train), greater_is_better=True)

# Example parameter grid
param_dict = {'alpha': np.linspace(1, 100, 100)}

# Set up and run GridSearchCV
model = GridSearchCV(estimator=Ridge(), param_grid=param_dict, scoring=custom_scorer)
model.fit(X_train, y_train)

# Best parameters and score
print("Best parameters:", model.best_params_)
print("Best score:", model.best_score_)
```

```python
model.best_params_, model.best_score_
```

```python
# Print the cross-validation results for all parameters
print("CV Results (mean scores across folds):")
print(model.cv_results_['mean_test_score'])  # Mean score across folds for each parameter

# Print all fold scores for each parameter
print("\nCV Results (per-fold scores):")
print(model.cv_results_['split0_test_score'])  # Scores for fold 1
print(model.cv_results_['split1_test_score'])  # Scores for fold 2
print(model.cv_results_['split2_test_score'])  # Scores for fold 3
print(model.cv_results_['split3_test_score'])  # Scores for fold 4
print(model.cv_results_['split4_test_score'])  # Scores for fold 5

# Best parameters and score
print("\nBest parameters:", model.best_params_)
print("Best mean score across all folds:", model.best_score_)
```
